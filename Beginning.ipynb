```python
bright_pic ="gdrive/MyDrive/Figure Layout and Classification Independent Study/data/ScannedPages/1960ApJ___131__282B_p1.jpeg"
bright = cv.imread(bright_pic)
plt.imshow(bright)
```

```python
hist,bins = np.histogram(bright.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(bright.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.yscale("log")
plt.show()
```

```python
#have to choose a brighter image to see results like the website
```

```python
cdf_m = np.ma.masked_equal(cdf,0)
cdf_m = (cdf_m - cdf_m.min())*255/(cdf_m.max()-cdf_m.min()) #this is an equation to calc. the histogram equalization equation
cdf = np.ma.filled(cdf_m,0).astype('uint8')
```

```python
img2 = cdf[bright]
plt.imshow(img2)
```
