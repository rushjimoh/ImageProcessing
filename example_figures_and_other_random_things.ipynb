````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# making a fake figure
fake_img1 = np.random.random((3300,2500))
fake_img2 = np.random.random((3300,2500))
````

````python
plt.imshow(fake_img1)
````

One way:

````python
fig, ax = plt.subplots(1,2)
````

````python
ax
````

````python
fig, ax = plt.subplots(1,2)

ax[0].imshow(fake_img1)
ax[0].set_title('Fake image 1')

ax[1].imshow(fake_img2)
ax[1].set_title('Fake image 2')

plt.show()
````

````python
fig, ax = plt.subplots(1,2, figsize=(20,30))

ax[0].imshow(fake_img1)
ax[0].set_title('Fake image 1')

ax[1].imshow(fake_img2)
ax[1].set_title('Fake image 2')

plt.show()
````

````python
fig, ax = plt.subplots(2,1, figsize=(5,8))

ax[0].imshow(fake_img1)
ax[0].set_title('Fake image 1')

ax[1].imshow(fake_img2)
ax[1].set_title('Fake image 2')

plt.show()
````

Another option:

https://stackoverflow.com/questions/14770735/how-do-i-change-the-figure-size-with-subplotsf

```
f = plt.figure(figsize=(10,3))
ax = f.add_subplot(121)
ax2 = f.add_subplot(122)
x = np.linspace(0,4,1000)
ax.plot(x, np.sin(x))
ax2.plot(x, np.cos(x), 'r:')
```

````python
plt.subplot(121); plt.imshow(fake_img1),plt.title('Fake image 1')
plt.xticks([]), plt.yticks([]) # no x/y axis
plt.subplot(122),plt.imshow(fake_img2),plt.title('Fake image 2')
plt.xticks([]), plt.yticks([])
plt.show()
````

````python
f = plt.figure(figsize=(20,30))
ax = f.add_subplot(121)
ax2 = f.add_subplot(122)

# first image
ax.imshow(fake_img1),ax.set_title('Fake image 1')
ax.set_xticks([]), ax.set_yticks([]) # no x/y axis

# second image
ax2.imshow(fake_img2),ax2.set_title('Fake image 2')
ax2.set_xticks([]), ax2.set_yticks([])
plt.show()
````
